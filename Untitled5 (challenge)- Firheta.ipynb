```python
# challenge 1
import random as rand
print("""what is your choice?
1. rock 
2. scissor 
3. paper """)

def game(arg, random_number):
    if arg == 1 and random_number == 2:
        message = "you win"
    elif arg == 1 and random_number == 3 :
        message = "you lose"
    elif arg == 2 and random_number == 1:
        message = "you lose"
    elif arg == 2 and random_number == 3:
        message = "you win"
    elif arg == 3 and random_number == 1:
        message = "you win"
    elif arg == 3 and random_number == 2:
        message = "you lose"
    else:
        message = "It's a draw"
    return message

while True:
    index = rand.randint(1, 4)
    ans = input("please enter number (or 'quit' to stop)")
    if ans== "quit":
        break
    print(game(int(ans), index))
print("Thank you for playing!")
```
